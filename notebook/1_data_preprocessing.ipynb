```python
import pandas as pd
import os

cols = [
'Severity',
'Start_Time',
'Temperature(F)',
'Humidity(%)',
'Visibility(mi)',
'Wind_Speed(mph)',
'Weather_Condition',
'Sunrise_Sunset'
]

df = pd.read_csv('../data/US_Accidents_March23.csv', usecols=cols, nrows=100000)

df['Start_Time'] = pd.to_datetime(df['Start_Time'])

df['hour'] = df['Start_Time'].dt.hour
df['day_of_week'] = df['Start_Time'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

df = df.drop(columns=['Start_Time'])

df = df.dropna()

df = pd.get_dummies(df, drop_first=True)

os.makedirs('../data', exist_ok=True)

df.to_csv('../data/accidents_processed.csv', index=False)


print(df.shape)
```
