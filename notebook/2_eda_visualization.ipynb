```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv('../data/accidents_processed.csv')

os.makedirs('../results', exist_ok=True)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='hour', data=df)
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.savefig('../results/accidents_by_hour.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='day_of_week', data=df)
plt.title("Accidents by Day of Week (0=Mon, 6=Sun)")
plt.xlabel("Day of Week")
plt.ylabel("Number of Accidents")
plt.savefig('../results/accidents_by_day.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df.groupby('hour').size().plot(kind='line', marker='o')
plt.title("Accident Trend by Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.savefig('../results/accidents_hour_trend.png', bbox_inches='tight')
plt.show()
```

```python
weather_cols = [col for col in df.columns if "Weather_Condition" in col]

weather_counts = df[weather_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=weather_counts.values[:10], y=weather_counts.index[:10])
plt.title("Top Weather Conditions During Accidents")
plt.xlabel("Number of Accidents")
plt.savefig('../results/weather_accidents.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm', center=0)
plt.title("Feature Correlation Heatmap")
plt.savefig('../results/correlation_heatmap.png', bbox_inches='tight')
plt.show()
```
